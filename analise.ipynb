## Dicionário de Dados

---




### Como Ler:

- **Usuário (`user_id`)**: Identificador único de um usuário no e-commerce.
- **Sessão (`user_session`)**: Identificador temporário de sessão de um usuário, onde podem ocorrer múltiplas ações.
- **Evento (`event_type`)**: Ação realizada pelo usuário, podendo ser adição ao carrinho, visualização, remoção ou compra de um produto.
- **Produto (`product_id`)**: Identificador único do produto envolvido no evento.
- **Marca (`brand`)**: Nome da marca do produto, em letras minúsculas. Pode estar ausente.
- **Categoria (`category_code`)**: Taxonomia da categoria do produto. Pode estar ausente.
- **Preço (`price`)**: Valor do produto no momento do evento. Sempre presente.

### Estrutura de Arquivo:

- **`event_time`**: Hora do evento (UTC).
- **`event_type`**: Tipo de evento (`view`, `cart`, `remove_from_cart`, `purchase`).
- **`product_id`**: ID do produto.
- **`category_id`**: ID da categoria do produto.
- **`category_code`**: Taxonomia da categoria do produto.
- **`brand`**: Nome da marca.
- **`price`**: Preço do produto.
- **`user_id`**: ID do usuário.
- **`user_session`**: ID da sessão do usuário.

### Tipos de Eventos:

- **`view`**: O usuário visualizou um produto.
- **`cart`**: O usuário adicionou um produto ao carrinho.
- **`remove_from_cart`**: O usuário removeu um produto do carrinho.
- **`purchase`**: O usuário comprou um produto.

### Várias Compras por Sessão:

Note que uma sessão de usuário pode incluir múltiplos eventos de `purchase`, que são considerados parte de um único pedido. Estes refletem o comportamento normal do usuário em uma sessão de compras.

### Comportamento (eventos) por mês

```python
df_ecommerce_tratado['month'] = df_ecommerce['event_time'].dt.month_name()
months = df_ecommerce_tratado['month'].unique()  # Isso deve conter os nomes dos meses

# Configurações de plotagem
sns.set(style="whitegrid")
plt.figure(figsize=(20, 16))

# Loop para criar os subplots para cada mês
for i, month in enumerate(months, 1):
    plt.subplot(3, 2, i)  # Ajuste as dimensões conforme necessário
    month_data = df_ecommerce_tratado[df_ecommerce_tratado['month'] == month]
    sns.countplot(data=month_data, x='event_type', palette="Set2")
    plt.title(f'Distribution of event type - {month}')
    plt.xlabel('Event Type')
    plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

### Tendência da Frequência (eventos)

```python
# Supondo que df_ecommerce já esteja com a coluna 'event_time' no formato datetime.
# Primeiro, adicionamos uma coluna que representa o mês e o ano para agrupar os dados.
df_ecommerce_tratado['month_year'] = df_ecommerce_tratado['event_time'].dt.to_period('M')

# Agora agrupamos os dados por mês/ano e tipo de evento, e contamos as ocorrências.
trend_data = df_ecommerce_tratado.groupby(['month_year', 'event_type']).size().reset_index(name='Freq')

# Convertendo 'month_year' para string para facilitar a plotagem.
trend_data['month_year'] = trend_data['month_year'].astype(str)

# Pivotear os dados para o formato largo para facilitar a plotagem com Seaborn ou Matplotlib.
trend_data_pivot = trend_data.pivot(index='month_year', columns='event_type', values='Freq').fillna(0)

# Plotando as tendências com Seaborn.
plt.figure(figsize=(14, 7))
sns.lineplot(data=trend_data_pivot)
plt.title('Event History Trend Over Months')
plt.xlabel('Month-Year')
plt.ylabel('Frequency of Events')
plt.xticks(rotation=45)
plt.legend(title='Event Type')
plt.tight_layout()
plt.show()
```

### Análise de Vendas ao Longo do Tempo

```python
# Filtro para compras e seleção das colunas relevantes
purchases = df_ecommerce_tratado[df_ecommerce_tratado['event_type'] == 'purchase'][['price', 'event_date']]

# Agregação da soma de vendas diárias
daily_sales = purchases.groupby('event_date').agg(total_sales=('price', 'sum')).reset_index()

# Conversão da coluna event_date para datetime (apenas por garantia)
daily_sales['event_date'] = pd.to_datetime(daily_sales['event_date'])

# Ordenação dos dados por data
daily_sales = daily_sales.sort_values('event_date')

# Encontrando os dias com vendas máximas e mínimas
max_sales_day = daily_sales[daily_sales['total_sales'] == daily_sales['total_sales'].max()]
min_sales_day = daily_sales[daily_sales['total_sales'] == daily_sales['total_sales'].min()]

# Plotagem das vendas diárias
plt.figure(figsize=(10, 6))
sns.lineplot(data=daily_sales, x='event_date', y='total_sales', label='Total Sales')

# Adicionar linhas horizontais para vendas máximas e mínimas
plt.axhline(y=daily_sales['total_sales'].max(), color='red', linestyle='--', label='Max Sales')
plt.axhline(y=daily_sales['total_sales'].min(), color='green', linestyle='--', label='Min Sales')

# Adicionando título e rótulos
plt.title('Total Sales Across Time Period')
plt.xlabel('Time Period')
plt.ylabel('Total Sales')

# Exibição do gráfico
plt.legend()
plt.show()

# Exibir os dias com vendas máximas e mínimas
print(max_sales_day, min_sales_day)
```

## Impacto na taxa de conversão e abandono de carrinho.
---

A taxa de abandono de carrinho é a porcentagem geral de consumidores que adicionaram itens ao carrinho, mas o abandonaram antes de finalizar a compra, representando clientes potenciais que estavam quase concluindo a compra, mas desistiram por algum motivo. Essa taxa indica o quão promissor poderia ser o processo de finalização de compra. A taxa de conversão é a porcentagem geral de visitantes que acessaram o site e efetuaram uma compra com sucesso. 

A partir da análise seguinte, há um claro aumento nas taxas de conversão (de 6% para 6,9%) e uma diminuição na taxa de abandono de carrinho (de 80% para 72,5%) de outubro a novembro de 2019. A tendência parece mudar drasticamente após novembro. A taxa de abandono de carrinho aumentou nos meses seguintes, mas não ultrapassou a marca alta de 80%, ficando em 78% em fevereiro de 2020, enquanto a taxa de conversão caiu abaixo do ponto mais baixo para pouco acima de 5,75%. 

A tendência geral de taxas de conversão acima da média pode ser devido a vários motivos, um dos quais poderia ser que empresas de médio porte trazem menos tráfego, mas clientes relativamente mais fiéis. 

Portanto, nosso objetivo deve ser diminuir a taxa de abandono de carrinho, que tem um efeito indireto nas taxas de conversão. Esta alta taxa pode indicar uma experiência menos fluida no processo de check-out, taxa de envio alta, promoções de remarketing fracas ou a falta de uma função de check-out para visitantes.

### Taxa de abandono do carrinho por mês

```python
# Suponha que 'trend_data' seja o seu DataFrame que já contém os dados de eventos.

# Filtrar os DataFrames por tipo de evento
cart = trend_data[trend_data['event_type'] == 'cart']
purchase = trend_data[trend_data['event_type'] == 'purchase']
removed = trend_data[trend_data['event_type'] == 'remove_from_cart']
view = trend_data[trend_data['event_type'] == 'view']

# Merge por mês e eventos
merged = pd.merge(cart, purchase, on='month_year', suffixes=('_cart', '_purchase'))
merged = pd.merge(merged, removed, on='month_year')
merged = pd.merge(merged, view, on='month_year', suffixes=('_removed', '_view'))

# Calcular a taxa de abandono do carrinho e a taxa de conversão
merged['ab_rate'] = (1 - (merged['Freq_purchase'] / merged['Freq_cart'])) * 100
merged['conv_rate'] = (merged['Freq_purchase'] / (merged['Freq_cart'] + 
                                                  merged['Freq_removed'] + 
                                                  merged['Freq_view'])) * 100

# Selecionar as colunas necessárias para o plot
rate = merged[['month_year', 'ab_rate', 'conv_rate']]

# Plotando a taxa de abandono do carrinho e a taxa de conversão
fig, axes = plt.subplots(2, 1, figsize=(13, 10))
sns.lineplot(ax=axes[0], data=rate, x='month_year', y='ab_rate', marker='o')
axes[0].set_title('Cart Abandonment Rate per Month')
axes[0].set_xlabel('Time Period')
axes[0].set_ylabel('Abandonment Rate (in %)')

sns.lineplot(ax=axes[1], data=rate, x='month_year', y='conv_rate', marker='o', color='orange')
axes[1].set_title('Conversion Rate per Month')
axes[1].set_xlabel('Time Period')
axes[1].set_ylabel('Conversion Rate (in %)')

plt.tight_layout()
plt.show()
```

Os 10 principais clientes que geraram a maior receita são destacados. Ao longo dos 5 meses, trouxeram mais de 3500 de receita. Em média, os clientes estão dispostos a gastar mais de 1000 dolares (ou moeda local) ao longo de 5 meses, sugerindo um bom poder de retenção dentro da empresa. ### Comportamento usuário por hora do Dia

De acordo com a análise  os clientes geralmente visualizam produtos de valor alto, mas as compras sempre ocorrem dentro dos produtos de valor mais baixo. Isso indica o que um cliente olha, mas não consegue comprar devido a fatores que indicam um preço de venda alto. Isso sugere ou que esses produtos estão extremamente caros. Este produto visualizado tem potencial para trazer taxas de conversão e retenção mais altas se oferecido de maneira atraente em promoções.

Além disso, é surpreendente ver que a maioria dos produtos é visualizada, removida do carrinho e adicionada ao carrinho por volta das 19h, enquanto a compra é feita durante as 11h, sugerindo que as horas da manhã, especialmente durante os dias de semana, são reservadas apenas para a compra do produto e não para outras atividades do funil de vendas.

```python
# Verifique se 'event_time' é do tipo datetime e extraia a hora (garantia).
df_ecommerce_tratado['hour'] = df_ecommerce_tratado['event_time'].dt.hour

# Agregue a soma dos preços por tipo de evento e hora.
time_hour = df_ecommerce_tratado.groupby(['event_type', 'hour'])['price'].sum().reset_index()

# Normalizando os preços.
time_hour['scaled_price'] = (time_hour['price'] - time_hour['price'].mean()) / time_hour['price'].std()

# Gráfico de linha da relação preço-hora por tipo de evento.
plt.figure(figsize=(10, 6))
sns.lineplot(data=time_hour, x='hour', y='scaled_price', hue='event_type', lw=1.2)

# Linhas verticais para horas de pico.
plt.axvline(x=19, linestyle='--', color='black', label='Peak - View, Cart, Remove')
plt.axvline(x=11, linestyle='--', color='black', label='Peak - Purchase')

# Títulos e rótulos.
plt.title('Relationship between Price and Hour of the Day Based on Event Type')
plt.xlabel('Time (in Hours)')
plt.ylabel('Scaled Price')
plt.xticks(range(0, 25))
plt.legend()
plt.tight_layout()
plt.show()
```
